# Text processing 

```python
zomato_filled_categoricaldata['lowercase']=zomato_filled_categoricaldata['review_description'].apply(lambda x:' '.join(x.lower() for x in x.split()))
```

```python
zomato_filled_categoricaldata
```

```python
import string 
string.punctuation
```

```python
zomato_filled_categoricaldata['cleanedpunctuation']=zomato_filled_categoricaldata['lowercase'].str.replace('!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~',"")
```

```python
zomato_filled_categoricaldata
```

```python
def punctuation(text):
    word="".join([word for word in text if word not in string.punctuation])
    return word
```

```python
zomato_filled_categoricaldata['punctuation_removed']=''
zomato_filled_categoricaldata['punctuation_removed']=zomato_filled_categoricaldata['cleanedpunctuation'].apply(lambda text:(punctuation(text)))
```

```python
zomato_filled_categoricaldata
```

```python
zomato_filled_categoricaldata['numbers removed']=zomato_filled_categoricaldata['punctuation_removed'].str.replace("\d","")
```

```python
zomato_filled_categoricaldata
```

```python
import nltk 
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
n=stopwords.words('english')
n
```

```python
zomato_filled_categoricaldata['cleaned with stopwords']=zomato_filled_categoricaldata['numbers removed'].apply(lambda x:' '.join(word for word in x.split() if word not in n))
```

```python
zomato_filled_categoricaldata
```

```python
import re
def remove_emojis(text):
    emoji_pattern=re.compile("[" 
                             u"\U0001F600-\U0001F64F"
                             u"\U0001F300-\U0001F5FF"
                             u"\U0001F680-\U0001F6FF"
                             u"\U0001F1E0-\U0001F1FF"
                             u"\U00002702-\U000027B0"
                             u"\U000024C2-\U0001F251" 
                             "]+",flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)
```

```python
zomato_filled_categoricaldata['emojis_removed']=''
zomato_filled_categoricaldata['emojis_removed']=zomato_filled_categoricaldata['cleaned with stopwords'].apply(lambda text:(remove_emojis(text)))
```

```python
zomato_filled_categoricaldata
```

# now till here text processing like converting small letters,removing punctuations,removing numbers,stop words,emojis done.

```python
zomato_concat=pd.concat([zomato_filled_numericaldata[:5000],zomato_filled_categoricaldata[:5000]],axis=1)
zomato_concat
```

```python
from nltk import word_tokenize
```

```python
n=zomato_concat[zomato_concat['rating']==5]
ratings=""
total_words= []
for rates in n['cleaned with stopwords']:
    ratings+= rates
    words=word_tokenize(rates)
    for word in words:
        if word not in set(stopwords.words('english')):
            total_words.append(word)
```

```python
x=zomato_filled_categoricaldata['emojis_removed'][:50000].values
x
```

```python
y=zomato_filled_numericaldata['rating'][:50000].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=12)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
v=TfidfVectorizer()
```

```python
X_train=v.fit_transform(x_train,y_train)
```

```python
from collections import Counter
Counter(y_train)
```

# This is Balanced data

```python
X_test=v.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
r=RandomForestClassifier()
```

```python
r.fit(X_train,y_train)
```

```python
predict=r.predict(X_test)
predict
```

```python
r.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
r=pd.DataFrame({'actual':y_test,'predicted':predict})
r
```

```python
r['predicted'].min()
```

```python
print(classification_report(y_test,predict,digits=3))
```
